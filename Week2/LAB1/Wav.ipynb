```python
import scipy as sp
from scipy.fftpack import fft
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

```python
s = wavfile.read('score.WAV')
print(s)
#wavfile.sys.getsizeof(s)
print((s[1][10000:15000]))
```

```python
freq = s[0]
n = 512
t = fft(s[1],n)
print(len(t))
```

```python
x = range(len(s[1]))
ft = sp.arange(0,freq,freq/n)
plt.plot(x,s[1],'r')
```

```python
print(len(ft))
print(len(t))
plt.plot(ft,abs(t),'b')
```
