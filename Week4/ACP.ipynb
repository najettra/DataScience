```python
import numpy as np
import random
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
data = []
for i in range(500):
    x = random.normalvariate(10,4)
    y = random.normalvariate(10,4)
    z = random.normalvariate(10,4)
    data.append([x,y,z])
    
```

```python
print(len(data))


for i in range(500):
    plt.scatter(data[i][0],data[i][1],data[i][2])
   
```

```python
for i in range(100):
    
    plt.scatter(i,data[i][2])
```

```python
scale_data = []
for i in range(100):
    scale_data.append([(data[i][0]-10)/2,(data[i][1]-10)/2,(data[i][2]-10)/2])
    plt.scatter(i,(data[i][2]-10)/2)
```

### Covariance matrix

```python
#scale_data = preprocessing.scale(scale_data)
mat_cov= np.cov(np.array(data).T)
```

### Correlation matrix

```python
mat_corr= np.corrcoef(np.array(scale_data).T)
```
